## In-silico mutagenesis of the CACNA1I gene

Here, we plot the results for the in-silico mutagenesis of CACNA1I (Figure 4EF, S11-13). In the tutorials folder, there is a notebook explaining how to generate results for any other gene for the ISM. 


```python
import os
import warnings

import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle
import seaborn as sns
import pickle

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
cp = pd.read_csv('../../Zenodo/human/M1/subclass/subclass.csv', index_col=0).columns
ct_mapping = np.vstack([cp, cp, cp])
ct_mapping[0,:] = 'All'
ct_mapping[1,[0,1,11,12,13,18]] = 'Non-neuronal'
ct_mapping[1,[10,14,15,16,17,19]] = 'GABAergic'
ct_mapping[1,[2,3,4,5,6,7,8,9]] = 'Glutamatergic'
```

### Read the effect of all predictions

```python
with open('../../Zenodo/human/M1/CACNA1I_subclass.pickle', 'rb') as f:
    sc = pickle.load(f)
```

### Plot mutation profile for Sst Chodl

Figure 4E

```python
sst_ch = sc[:,17]
sst_ch_max = np.max(sst_ch,axis=1)
sst_ch_min = np.min(sst_ch,axis=1)
```

```python
offset = 39563753 #Position on the chromosome of first input base

fig = plt.figure(figsize=(10, 1.25))  
ax = plt.subplot()

# Plot minimum and maximum effect
plt.plot(np.linspace(0, 10499, num=10500)+offset, sst_ch_max, rasterized=True, color=sns.color_palette('colorblind')[0])
plt.plot(np.linspace(0, 10499, num=10500)+offset, sst_ch_min, rasterized=True, color=sns.color_palette('colorblind')[1])

# Add the candidate cis-regulatory elements (downloaded from SCREEN v3)
facecolor=sns.color_palette('colorblind')[7]
edgecolor='none'
alpha=0.5

#EH38E3482229
#EH38E3482230
#EH38E3482231
#EH38E3482232
#EH38E2164864
elements = [[39567387, 39567737],
            [39568602, 39568948],
            [39569441, 39569761],
            [39569835, 39570117],
            [39570245, 39570564]]

rec = [Rectangle((x1, -0.135), x2-x1, 0.025) for x1, x2 in elements]

pc = PatchCollection(rec, facecolor=facecolor, alpha=alpha,
                         edgecolor=edgecolor)
ax.add_collection(pc)

plt.ticklabel_format(useOffset=False, style='plain')
plt.title('CACNA1I')
plt.ylabel('Predicted effect')

# Location we're going to zoom in on
plt.axvline(39570200, color='k', alpha=0.5)
plt.axvline(39571500, color='k', alpha=0.5)

plt.text(0.01, 0.95, 'Sst Chodl',
         verticalalignment='top', horizontalalignment='left',
         transform=ax.transAxes, fontsize=12)

sns.despine()

plt.xlim([39563753, 39574253])
plt.ylim([-0.15, 0.15])
plt.xticks([39564000, 39566000, 39568000, 39570000, 39572000, 39574000], 
           [39564000, 39566000, 39568000, 39570000, 39572000, 39574000])

plt.show()
```

### Plot difference between Sst Chodl and L2/3 IT

Figure S11

```python
L23 = sc[:,2]

diff = sst_ch - L23
```

In Figure S11, we only plot mutations where both predict an increase in expression or a decrease. Here, we check if we can indeed ignore the positions where the sign changes. In the histogram, you see that it almost never happens that one predicts a positive effect and one a negative. If it happens, the effect is also small, so we can ignore these.


```python
plt.hist(sst_ch[np.where((sst_ch < 0) & (L23 > 0))])
plt.show()

plt.hist(sst_ch[np.where((sst_ch > 0) & (L23 < 0))])
plt.show()
```

```python
## Decrease in expression (Figure S11A)
diff = L23 - sst_ch
xx, yy = np.where((sst_ch >= 0) | (L23 >= 0))

for i, x in enumerate(xx):
    diff[x,yy[i]] = 0
    
min_diff = np.min(diff, axis=1)
max_diff = np.max(diff, axis=1)
min_diff[min_diff > 0] = 0
max_diff[max_diff < 0] = 0

fig = plt.figure(figsize=(12, 1.5))  
ax = plt.subplot()

plt.plot(np.linspace(0, 10499, num=10500)+offset, min_diff, rasterized=True, color=sns.color_palette('colorblind')[0])
plt.plot(np.linspace(0, 10499, num=10500)+offset, max_diff, rasterized=True, color=sns.color_palette('colorblind')[1])

elements = [[39567387, 39567737],
            [39568602, 39568948],
            [39569441, 39569761],
            [39569835, 39570117],
            [39570245, 39570564]]

rec = [Rectangle((x1, -0.02), x2-x1, 0.01) for x1, x2 in elements]

pc = PatchCollection(rec, facecolor=facecolor, alpha=alpha,
                         edgecolor=edgecolor)
ax.add_collection(pc)

sns.despine()

plt.xlim([39563753, 39574253])
plt.ylim([-0.023, 0.047])
plt.xticks([39564000, 39566000, 39568000, 39570000, 39572000, 39574000], 
           [39564000, 39566000, 39568000, 39570000, 39572000, 39574000])

plt.title('Decrease in expression')

plt.show()
```

```python
## Increase in expression (Figure S11B)
diff = sst_ch - L23
xx, yy = np.where((sst_ch <= 0) | (L23 <= 0))

for i, x in enumerate(xx):
    diff[x,yy[i]] = 0
    
min_diff = np.min(diff, axis=1)
max_diff = np.max(diff, axis=1)
min_diff[min_diff > 0] = 0
max_diff[max_diff < 0] = 0

fig = plt.figure(figsize=(12, 1.5))  
ax = plt.subplot()

plt.plot(np.linspace(0, 10499, num=10500)+offset, min_diff, rasterized=True, color=sns.color_palette('colorblind')[0])
plt.plot(np.linspace(0, 10499, num=10500)+offset, max_diff, rasterized=True, color=sns.color_palette('colorblind')[1])

elements = [[39567387, 39567737],
            [39568602, 39568948],
            [39569441, 39569761],
            [39569835, 39570117],
            [39570245, 39570564]]

rec = [Rectangle((x1, -0.02), x2-x1, 0.01) for x1, x2 in elements]

pc = PatchCollection(rec, facecolor=facecolor, alpha=alpha,
                         edgecolor=edgecolor)
ax.add_collection(pc)


sns.despine()

plt.xlim([39563753, 39574253])
plt.ylim([-0.023, 0.047])
plt.xticks([39564000, 39566000, 39568000, 39570000, 39572000, 39574000], 
           [39564000, 39566000, 39568000, 39570000, 39572000, 39574000])

plt.title('Increase in expression')

plt.show()
```

### Predicted effect of GWAS variants

```python
# Figure 4F
GWAS1 = sc[6789,:,1:]
GWAS1 = pd.DataFrame(GWAS1, index=cp, columns=['C', 'T', 'G'])
GWAS1['class'] = ct_mapping[1]
GWAS1.reset_index(inplace=True)
GWAS1 = GWAS1.rename(columns = {'index':'subclass'})
GWAS1 = pd.melt(GWAS1, id_vars=['class', 'subclass'])

plt.figure(figsize=(3.5, 2))
sns.swarmplot(data=GWAS1, y='variable', x='value', hue = 'class', size=5, palette='colorblind')
sns.despine()
plt.xlim([0, 0.062])
plt.ylabel('')
plt.xlabel('Predicted effect')
plt.title('rs7288455 (CACNA1I)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
